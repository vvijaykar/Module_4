```python
import pandas as pd
import numpy as np
```

```python
purchase_csv = "Data/purchase_data_new.csv"

purchase_data = pd.read_csv(purchase_csv)
```

```python
# Player Count
    # Display the Total Player Count

total_players = len(purchase_data["SN"].value_counts())
                
num_players = pd.DataFrame({"Total Players":[total_players]})
                 
num_players.head()
```

```python
# Purchasing Analysis
    # Run basic calculations to obtain number of unique items, average price, etc.
    # Create a summary data frame to hold the results
    # Optional: give the displayed data cleaner formatting
    # Display the summary data frame

unique_items = len(purchase_data["Item ID"].unique())

average_price = (purchase_data["Price"].sum())/len(purchase_data["Purchase ID"])

num_purchase = len(purchase_data["SN"])

total_rev = (purchase_data["Price"].sum())

summary_table = pd.DataFrame({"Number of Unique Items":[unique_items], 
                             "Average Price":[average_price],
                             "Number of Purchases": [num_purchase],
                             "Total Revenue" : [total_rev]})

summary_table.style.format({"Average Price":"${:,.2f}",
                         "Total Revenue": '${:,.2f}'})
```

```python
# Gender Demographics
    # Percentage and Count of Male Players
    # Percentage and Count of Female Players
    # Percentage and Count of Other / Non-Disclosed

#gender_demograph = purchase_data.groupby("Gender")

gender_demograph = purchase_data["Gender"].value_counts()


#gender_count_total = gender_demograph["SN"].nunique()

players_percentage = (gender_count_total/total_players) * 100 


gender_table = pd.DataFrame({"Total Count": gender_count_total,
                            "Percentage of Players": players_percentage})

gender_table.style.format({"Percentage of Players": '{:,.2f}'})

gender_table.sort_values(["Total Count"], ascending = False)

gender_table.index.name = None

gender_table
```

```python
# Purchasing Analysis (Gender)
    # Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
    # Create a summary data frame to hold the results
    # Optional: give the displayed data cleaner formatting
# Display the summary data frame


purchase_count = gender_demograph["Purchase ID"].count()

avg_purchase = gender_demograph["Price"].mean()

avg_purchase_total = gender_demograph["Price"].sum()

avg_purchase_per_person = avg_purchase_total/gender_count_total

gender_analysis = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": avg_purchase,
                                    "Average Purchase Value":avg_purchase_total,
                                    "Avg Purchase Total per Person": avg_purchase_per_person})

gender_analysis.style.format({"Average Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})

gender_analysis.index.name = "Gender"

gender_analysis
```

```python
# Age Demographics

age_groups = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_groups, labels=group_names)
purchase_data
```

```python
# Age Demographics

age_grouped = purchase_data.groupby("Age Group")

total_count_age = age_grouped["SN"].nunique()

age_percentage = (total_count_age/total_players) * 100

age_table = pd.DataFrame({"Percentage of Players": age_percentage, "Total Count": total_count_age})

age_table.index.name = None

age_table.style.format({"Percentage of Players":"{:,.2f}"})
```

```python
# Purchasing Analysis (Age)

purchase_by_age = age_grouped["Purchase ID"].count()
avg_purchase_age = age_grouped["Price"].mean()
total_purchase_value = age_grouped["Price"].sum()
avg_purchase_per_person_age = total_purchase_value/total_count_age


age_demograph = pd.DataFrame({"Purchase Count": purchase_by_age,
                                 "Average Purchase Price": avg_purchase_age,
                                 "Total Purchase Value":total_purchase_value,
                                 "Average Purchase Total per Person": avg_purchase_per_person_age})

age_demograph.index.name = None

age_demograph.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

```python
# Top Spenders

spender_metrics = purchase_data.groupby("SN")
purchase_count_spender = spender_metrics["Purchase ID"].count()

avg_purchase_price_spender = spender_metrics["Price"].mean()

purchase_total_spender = spender_metrics["Price"].sum()

top_spenders = pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price": avg_purchase_price_spender,
                             "Total Purchase Value":purchase_total_spender})

formatted_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

formatted_spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

```python
# Most Popular Items

items = purchase_data[["Item ID", "Item Name", "Price"]]
item_metrics = items.groupby(["Item ID","Item Name"])
purchase_count_item = item_metrics["Price"].count()
purchase_value = (item_metrics["Price"].sum()) 
item_price = purchase_value/purchase_count_item

most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":purchase_value})

most_popular_organized = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()

most_popular_organized.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```

```python
# Most Profitable 

most_popular_organized = most_popular_items.sort_values(["Total Purchase Value"],
                                                   ascending=False).head()
most_popular_organized.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
